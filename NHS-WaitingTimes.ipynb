```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Make sure 'appointment_month' is in datetime format
national['appointment_month'] = pd.to_datetime(national['appointment_month'])

# Extract year and month for filtering and grouping
national['year'] = national['appointment_month'].dt.year
national['month'] = national['appointment_month'].dt.strftime('%b')  # abbreviated month name
national['month_num'] = national['appointment_month'].dt.month       # for sorting

# Group data by year, month, and service_setting
grouped = national.groupby(['year', 'month', 'month_num', 'service_setting'])['count_of_appointments'].sum().reset_index()

# Sort by month number for correct order
grouped = grouped.sort_values(by='month_num')

# Split the data into 2021 and 2022
data_2021 = grouped[grouped['year'] == 2021]
data_2022 = grouped[grouped['year'] == 2022]

# Set plot style
sns.set_style('whitegrid')
sns.set(rc={'figure.figsize':(15, 6)})

# Plot for 2021
plt.figure(figsize=(15, 6))
sns.lineplot(data=data_2021, x='month', y='count_of_appointments', hue='service_setting', marker='o')
plt.title('Monthly Appointments by Service Setting - 2021')
plt.xlabel('Month')
plt.ylabel('Count of Appointments')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot for 2022
plt.figure(figsize=(15, 6))
sns.lineplot(data=data_2022, x='month', y='count_of_appointments', hue='service_setting', marker='o')
plt.title('Monthly Appointments by Service Setting - 2022')
plt.xlabel('Month')
plt.ylabel('Count of Appointments')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# break down appointment number per month in 2021 ans 2022

# Convert date field and extract relevant time features
regional['appointment_month'] = pd.to_datetime(regional['appointment_month'])
regional['month_name'] = regional['appointment_month'].dt.strftime('%B')

# Define status categories
attended_statuses = ['Attended']
dna_statuses = ['DNA']

# Function to map season from month
def get_season(date):
    month = date.month
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Add season column
regional['season'] = regional['appointment_month'].apply(get_season)

# Group and summarize appointment counts per season and month
summary = regional[regional['appointment_status'].isin(attended_statuses + dna_statuses)] \
    .groupby(['season', 'month_name', 'appointment_status'])['count_of_appointments'] \
    .sum() \
    .unstack(fill_value=0) \
    .reset_index()

# Sort month order for readability
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
summary['month_order'] = summary['month_name'].apply(lambda x: month_order.index(x))
summary = summary.sort_values(['season', 'month_order']).drop(columns='month_order')

# Display the grouped summary
summary
```

```python
national = pd.read_excel('national_categories.xlsx')
regional = pd.read_csv('appointments_regional.csv')

# Convert appointment date
national['appointment_month'] = pd.to_datetime(national['appointment_month'])

# Filter for NHS North West London ICB - W2U3Z and date range
filtered = national[
    (national['icb_ons_code'] == 'W2U3Z') &
    (national['appointment_month'] >= '2022-01-01') &
    (national['appointment_month'] <= '2022-06-01') &
    (regional['appointment_status'] == 'DNA')  # Only missed appointments
]

# Group by service setting and sum missed appointments
missed_by_service = filtered.groupby('service_setting')['count_of_appointments'].sum().sort_values(ascending=False)

# Count unique service settings in the filtered data
num_service_settings = filtered['service_setting'].nunique()

missed_by_service, num_service_settings
```
